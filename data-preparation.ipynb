#### Téléchargement du jeu de données
Le fichier zip contenant toutes les données fait 6GB
Les images sont rangées par catégories dans des dossiers


#### Imports

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

#### Affichage des images

```python
folder_path = '5eme dataset/images/churros'

# Obtenir la liste de tous les fichiers dans le dossier
files = os.listdir(folder_path)

# Trier la liste des fichiers
files.sort()

# Obtenir le chemin complet du premier fichier
first_image_path = os.path.join(folder_path, files[0])

# Lire l'image
img = mpimg.imread(first_image_path)

# Afficher l'image
plt.imshow(img)
plt.show()
print("Image shape: ", img.shape)
print("Image type: ", img.dtype)
```

#### La taille moyenne des images

```python
# Initialiser les variables pour stocker la somme des dimensions
total_width = 0
total_height = 0
num_images = 0

# Parcourir tous les fichiers dans le dossier
for file in os.listdir(folder_path):
    # Vérifier si le fichier est une image (se termine par '.jpg')
    if file.endswith('.jpg'):
        # Ouvrir l'image et obtenir ses dimensions
        img = Image.open(os.path.join(folder_path, file))
        width, height = img.size

        # Ajouter les dimensions à la somme totale
        total_width += width
        total_height += height

        # Compter le nombre d'images
        num_images += 1

# Calculer les dimensions moyennes
average_width = total_width / num_images
average_height = total_height / num_images

print(f"Average width: {average_width} pixels")
print(f"Average height: {average_height} pixels")
```

#### Combien y a t'il d'images dans ce dataset et quelle est leur taille moyenne ?

```python
folder_path = '5eme dataset/images'
total_width = 0
total_height = 0
first_num_images = 0

for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.jpg'):
            first_num_images += 1

            img = Image.open(os.path.join(root, file))
            width, height = img.size

            total_width += width
            total_height += height

# Calculer les dimensions moyennes
average_width = total_width / first_num_images
average_height = total_height / first_num_images

print(f"Total number of images: {first_num_images}")
print(f"Average width: {average_width} pixels")
print(f"Average height: {average_height} pixels")
```
